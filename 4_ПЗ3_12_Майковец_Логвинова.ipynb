##Загружаем библиотеки и вводим значения для переменных
v = номер двойки = 12

α = номер первого из двойки по списку = 17

γ = номер второго из двойки по списку = 18

μ = количество букв в ФИО одного участника двойки = 23

β = α + γ = 17 + 18 = 35

θ = γ = 18

y = v = 12

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root, fsolve
import sympy as sp

v = 12
alpha = 17
gamma = 18
mu = 23
beta = 35
theta = 18
y = 12

print("Параметры:")
print(f"v = {v}, α = {alpha}, γ = {gamma}, μ = {mu}, β = {beta}, θ = {theta}, y = {y}")
```

## Задание 1
Решить алгебраическое уравнение $ (𝜈 + 𝑥)^3 − 2𝑥^2−𝛼𝑥 − 17 = 0 $ с использованием функции root(). Получить графическое решение.

```python
def equation1(x):
    return (v + x)**3 - 2*x**2 - alpha*x - 17

# Графическое решение
x_vals = np.linspace(-10, 5, 400)
y_vals = equation1(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$(v + x)^3 - 2x^2 - \\alpha x - 17$')
plt.axhline(y=0, color='r', linestyle='--', alpha=0.3)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Задание 1: График функции')
plt.legend()
plt.show()

# Численное решение
solution1 = root(equation1, 0)  # начальное приближение x=0
print(f"Задание 1: x = {solution1.x[0]:.6f}")
print(f"Проверка: f(x) = {equation1(solution1.x[0]):.2e}")
```

## Задание 2
Найти корни алгебраического полинома $ 𝛼𝑥^3 − (𝛾 +𝜗)𝑥^2 − 𝜇𝑥 + 𝜈 = 0 $ с помощью функции fsolve(). Привести графическое решение задачи.

```python
def equation2(x):
    return alpha*x**3 - (y + theta)*x**2 - mu*x + v

# Графическое решение
x_vals = np.linspace(-3, 5, 400)
y_vals = equation2(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$\\alpha x^3 - (y + \\theta)x^2 - \\mu x + v$')
plt.axhline(y=0, color='r', linestyle='--', alpha=0.3)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Задание 2: График полинома')
plt.legend()
plt.show()

# Численное решение
solution2 = fsolve(equation2, [-2, 0, 4])
print("Задание 2: Корни полинома:")
for i, root_val in enumerate(solution2):
    print(f"  x{i+1} = {root_val:.6f}")
    print(f"  Проверка: f(x{i+1}) = {equation2(root_val):.2e}")
```

##Задание 3
Решить систему линейных уравнений с помощью
функции fsolve():
$$
\begin{cases}
3x_1 + x_2 + x_3 + \gamma x_4 = \alpha \\
x_1 - \mu x_2 + \nu x_3 + 4x_4 = \beta \\
-5x_1 - x_3 - 7x_4 = -5 \\
x_1 - 6x_2 + \alpha x_3 + 6x_4 = 3
\end{cases}
$$

```python
def system3(X):
    x1, x2, x3, x4 = X
    eq1 = 3*x1 + x2 + x3 + y*x4 - alpha
    eq2 = x1 - mu*x2 + v*x3 + 4*x4 - beta
    eq3 = -5*x1 - x2 - 7*x4 + 5
    eq4 = x1 - 6*x2 + alpha*x3 + 6*x4 - 3
    return [eq1, eq2, eq3, eq4]

# Решение системы
solution3 = fsolve(system3, [0, 0, 0, 0])
x1, x2, x3, x4 = solution3

print("Задание 3: Решение системы:")
print(f"x₁ = {x1:.6f}")
print(f"x₂ = {x2:.6f}")
print(f"x₃ = {x3:.6f}")
print(f"x₄ = {x4:.6f}")

# Проверка
check = system3(solution3)
print("Проверка (должны быть близки к 0):")
for i, val in enumerate(check):
    print(f"  Уравнение {i+1}: {val:.2e}")
```

## Задание 4
Решить систему линейных уравнений с помощью
функции root():
$$
\begin{cases}
\beta x_1 + \gamma x_2 + \nu x_3 + x_4 = 10\mu + \nu \\
x_1 - 3\mu x_2 + 2x_3 + 4\alpha x_4 = 9 + \gamma \\
-5x_1 + \nu x_2 - x_3 - 7x_4 = -5 \\
\nu x_1 - 6x_2 + 2x_3 + 6x_4 = \mu
\end{cases}
$$

```python
def system4(X):
    x1, x2, x3, x4 = X
    eq1 = beta*x1 + gamma*x2 + v*x3 + x4 - (10*mu + v)
    eq2 = x1 - 3*mu*x2 + 2*x3 + 4*alpha*x4 - (9 + gamma)
    eq3 = -5*x1 + v*x2 - x2 - 7*x4 + 5
    eq4 = v*x1 - 6*x2 + 2*x3 + 6*x4 - mu
    return [eq1, eq2, eq3, eq4]

# Решение системы
solution4 = root(system4, [0, 0, 0, 0])
x1, x2, x3, x4 = solution4.x

print("Задание 4: Решение системы:")
print(f"x₁ = {x1:.6f}")
print(f"x₂ = {x2:.6f}")
print(f"x₃ = {x3:.6f}")
print(f"x₄ = {x4:.6f}")

# Проверка
check = system4(solution4.x)
print("Проверка (близки ли к 0):")
for i, val in enumerate(check):
    print(f"  Уравнение {i+1}: {val:.2e}")
```

## Задание 5
Решить систему нелинейных уравнений. Получить
графическое решение:

#Система 1:

$$
\begin{cases}
3\ln x + y\sin x + y^2 = 12 \\
\frac{x}{y^2+1} + 2^x - 3y^2 = 22
\end{cases}
$$

```python
def system1(X):
    x, y_val = X
    eq1 = 3*np.log(abs(x)) + y_val*np.sin(x) + y_val**2 - 12
    eq2 = x/(y_val**2 + 1) + 2**x - 3*y_val**2 - 22
    return [eq1, eq2]

# Графическое решение системы 1
x_range1 = np.linspace(0.1, 5, 100)
y_range1 = np.linspace(-3, 3, 100)
X1, Y1 = np.meshgrid(x_range1, y_range1)

Z1_1 = 3*np.log(np.abs(X1)) + Y1*np.sin(X1) + Y1**2 - 12
Z1_2 = X1/(Y1**2 + 1) + 2**X1 - 3*Y1**2 - 22

# Численное решение
sol1 = fsolve(system1, [2, 1])

plt.figure(figsize=(10, 8))
plt.contour(X1, Y1, Z1_1, levels=[0], colors='red', linewidths=3, linestyles='-')
plt.contour(X1, Y1, Z1_2, levels=[0], colors='blue', linewidths=3, linestyles='--')
plt.plot(sol1[0], sol1[1], 'go', markersize=12, markerfacecolor='gold', markeredgecolor='black', label=f'Решение: ({sol1[0]:.3f}, {sol1[1]:.3f})')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Система 1: Графическое решение', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.axis('equal')
plt.tight_layout()
plt.show()

print("СИСТЕМА 1:")
print(f"Решение: x = {sol1[0]:.6f}, y = {sol1[1]:.6f}")
print(f"Проверка: f1 = {system1(sol1)[0]:.2e}, f2 = {system1(sol1)[1]:.2e}")
print()
```

#Система 2
$$
\begin{cases}
x^2 - y^3 + \frac{x}{\sin y + 2} + 2^x = 15 \\
x\cos^2 y + e^{x-y} - y^2 = 17
\end{cases}
$$

```python
def system2(X):
    x, y_val = X
    eq1 = x**2 - y_val**3 + x/(np.sin(y_val) + 2) + 2**x - 15
    eq2 = x*np.cos(y_val)**2 + np.exp(x - y_val) - y_val**2 - 17
    return [eq1, eq2]

# Графическое решение системы 2
x_range2 = np.linspace(0, 4, 100)
y_range2 = np.linspace(-2, 2, 100)
X2, Y2 = np.meshgrid(x_range2, y_range2)

Z2_1 = X2**2 - Y2**3 + X2/(np.sin(Y2) + 2) + 2**X2 - 15
Z2_2 = X2*np.cos(Y2)**2 + np.exp(X2 - Y2) - Y2**2 - 17

# Численное решение
sol2 = fsolve(system2, [2, 1])

plt.figure(figsize=(10, 8))
plt.contour(X2, Y2, Z2_1, levels=[0], colors='red', linewidths=3, linestyles='-')
plt.contour(X2, Y2, Z2_2, levels=[0], colors='blue', linewidths=3, linestyles='--')
plt.plot(sol2[0], sol2[1], 'go', markersize=12, markerfacecolor='gold', markeredgecolor='black', label=f'Решение: ({sol2[0]:.3f}, {sol2[1]:.3f})')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Система 2: Графическое решение', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.axis('equal')
plt.tight_layout()
plt.show()

print("СИСТЕМА 2:")
print(f"Решение: x = {sol2[0]:.6f}, y = {sol2[1]:.6f}")
print(f"Проверка: f1 = {system2(sol2)[0]:.2e}, f2 = {system2(sol2)[1]:.2e}")
print()
```

#Система 3
$$
\begin{cases}
\frac{x}{y^2 + 1} + 2x\cos y + e^{-x} + xy = 38 \\
\frac{3xy}{\sin x + 2} + \ln(|y| + 6) - x^2 = 5
\end{cases}
$$

```python
def system3(X):
    x, y_val = X
    eq1 = x/(y_val**2 + 1) + 2*x*np.cos(y_val) + np.exp(-x) + x*y_val - 38
    eq2 = 3*x*y_val/(np.sin(x) + 2) + np.log(np.abs(y_val) + 6) - x**2 - 5
    return [eq1, eq2]

# Графическое решение системы 3
x_range3 = np.linspace(1, 6, 100)
y_range3 = np.linspace(0, 4, 100)
X3, Y3 = np.meshgrid(x_range3, y_range3)

Z3_1 = X3/(Y3**2 + 1) + 2*X3*np.cos(Y3) + np.exp(-X3) + X3*Y3 - 38
Z3_2 = 3*X3*Y3/(np.sin(X3) + 2) + np.log(np.abs(Y3) + 6) - X3**2 - 5

# Численное решение
sol3 = fsolve(system3, [3, 2])

plt.figure(figsize=(10, 8))
plt.contour(X3, Y3, Z3_1, levels=[0], colors='red', linewidths=3, linestyles='-')
plt.contour(X3, Y3, Z3_2, levels=[0], colors='blue', linewidths=3, linestyles='--')
plt.plot(sol3[0], sol3[1], 'go', markersize=12, markerfacecolor='gold', markeredgecolor='black', label=f'Решение: ({sol3[0]:.3f}, {sol3[1]:.3f})')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Система 3: Графическое решение', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.axis('equal')
plt.tight_layout()
plt.show()

print("СИСТЕМА 3:")
print(f"Решение: x = {sol3[0]:.6f}, y = {sol3[1]:.6f}")
print(f"Проверка: f1 = {system3(sol3)[0]:.2e}, f2 = {system3(sol3)[1]:.2e}")
print()
```

#Система 4
$$
\begin{cases}
x^2 + y^2 = 5xy + 3 \\
\frac{5y}{x^2 + 1} + 4\cos x = 6
\end{cases}
$$

```python
def system4(X):
    x, y_val = X
    eq1 = x**2 + y_val**2 - 5*x*y_val - 3
    eq2 = 5*y_val/(x**2 + 1) + 4*np.cos(x) - 6
    return [eq1, eq2]

# Графическое решение системы 4
x_range4 = np.linspace(-2, 3, 100)
y_range4 = np.linspace(-2, 3, 100)
X4, Y4 = np.meshgrid(x_range4, y_range4)

Z4_1 = X4**2 + Y4**2 - 5*X4*Y4 - 3
Z4_2 = 5*Y4/(X4**2 + 1) + 4*np.cos(X4) - 6

# Численное решение
sol4 = fsolve(system4, [1, 1])

plt.figure(figsize=(10, 8))
plt.contour(X4, Y4, Z4_1, levels=[0], colors='red', linewidths=3, linestyles='-')
plt.contour(X4, Y4, Z4_2, levels=[0], colors='blue', linewidths=3, linestyles='--')
plt.plot(sol4[0], sol4[1], 'go', markersize=12, markerfacecolor='gold', markeredgecolor='black', label=f'Решение: ({sol4[0]:.3f}, {sol4[1]:.3f})')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Система 4: Графическое решение', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.axis('equal')
plt.tight_layout()
plt.show()

print("СИСТЕМА 4:")
print(f"Решение: x = {sol4[0]:.6f}, y = {sol4[1]:.6f}")
print(f"Проверка: f1 = {system4(sol4)[0]:.2e}, f2 = {system4(sol4)[1]:.2e}")
print()
```

#Система 6
$$
\begin{cases}
