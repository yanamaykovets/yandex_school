## Задание 1. Взяв шаг, равный одной десятой длины интервала интегрирования, вычислить определенный интеграл по обобщенным формулам трапеций и Симпсона. Тот же интеграл вычислить средствами Python с помощью одной из приведенных в пособии специальных функций.


### Интеграл 1:
$\int_1^5 \sqrt{9x - 1}  dx$

#### Аналитическое решение:
$\int \sqrt{9x - 1}  dx = \frac{2}{27}(9x - 1)^{3/2} + C$

$\int_1^5 \sqrt{9x - 1}  dx = \left[\frac{2}{27}(9x - 1)^{3/2}\right]_1^5 = \frac{2}{27}(44^{3/2}) - \frac{2}{27}(8^{3/2}) = \frac{2}{27}(44\sqrt{44}) - \frac{2}{27}(8\sqrt{8}) \approx \frac{2}{27}(291.6) - \frac{2}{27}(22.63) \approx 21.6 - 1.68 = 19.92$

#### Численные методы:

Параметры:
- Интервал: [1, 5]
- Длина интервала: 4
- Шаг: 0.4
- Количество узлов: 11

Формула трапеций:

$I_{trap} = \frac{h}{2} \left[f(x_0) + 2\sum_{i=1}^{n-1} f(x_i) + f(x_n)\right]$

Формула Симпсона:

$I_{simp} = \frac{h}{3} \left[f(x_0) + 4\sum_{i=1,3,5...}^{n-1} f(x_i) + 2\sum_{i=2,4,6...}^{n-2} f(x_i) + f(x_n)\right]$


```python
import numpy as np
from scipy import integrate
import math

# Функция для интегрирования
def f1(x):
    return np.sqrt(9*x - 1)

# Аналитическое решение
def analytical_f1(a, b):
    return (2/27) * (9*b - 1)**(3/2) - (2/27) * (9*a - 1)**(3/2)

print("=" * 60)
print("Вычисление интеграла: ∫₁⁵ √(9x - 1) dx")
print("=" * 60)

# Параметры интегрирования
a1, b1 = 1, 5
L1 = b1 - a1
h1 = L1 / 10
n1 = 11
x1 = np.linspace(a1, b1, n1)
y1 = f1(x1)

print(f"Параметры:")
print(f"Интервал: [{a1}, {b1}]")
print(f"Длина интервала: {L1}")
print(f"Шаг: {h1}")
print(f"Количество узлов: {n1}")
print()

# Метод трапеций
trap1 = (h1/2) * (y1[0] + 2*np.sum(y1[1:-1]) + y1[-1])

# Метод Симпсона
if n1 % 2 == 1:
    simp1 = (h1/3) * (y1[0] + 4*np.sum(y1[1:-1:2]) + 2*np.sum(y1[2:-1:2]) + y1[-1])
else:
    simp1 = "Требуется нечетное количество узлов"

# Scipy
scipy1, err1 = integrate.quad(f1, a1, b1)

print("РЕЗУЛЬТАТЫ:")
print(f"Аналитическое решение: {analytical_f1(a1, b1):.6f}")
print(f"Метод трапеций:        {trap1:.6f}")
print(f"Метод Симпсона:        {simp1:.6f}")
print(f"Scipy (quad):          {scipy1:.6f}")
print(f"Погрешность Scipy:     {err1:.2e}")

print("\n" + "="*50)
print("Таблица значений:")
print("="*50)
print("i\t x_i\t\t f(x_i)")
print("-" * 35)
for i in range(n1):
    print(f"{i}\t {x1[i]:.1f}\t\t {y1[i]:.6f}")

# Вывод формул
print("\n" + "="*50)
print("Использованные формулы:")
print("="*50)
print("Метод трапеций:")
print("I_trap = h/2 * [f(x₀) + 2∑f(x_i) + f(x_n)]")
print()
print("Метод Симпсона:")
print("I_simp = h/3 * [f(x₀) + 4∑f(x_нечет) + 2∑f(x_чет) + f(x_n)]")
```

### Интеграл 2:  
$\int_2^3(1-2x)^4dx$

#### Аналитическое решение:
$\int (1 - 2x)^4 dx = -\frac{1}{10}(1 - 2x)^5 + C$

$\int_2^3 (1 - 2x)^4 dx = \left[-\frac{1}{10}(1 - 2x)^5\right]_2^3 = -\frac{1}{10}(1 - 6)^5 + \frac{1}{10}(1 - 4)^5 = -\frac{1}{10}(-5)^5 + \frac{1}{10}(-3)^5 = -\frac{1}{10}(-3125) + \frac{1}{10}(-243) = 312.5 - 24.3 = 288.2$

#### Численные методы:

Параметры:
- Интервал: [2, 3]
- Длина интервала: 1
- Шаг: 0.1
- Количество узлов: 11

### Интеграл 3:
$\int_{-2}^{0} (x^5 - 3x^2) dx$

#### Аналитическое решение:
$\int (x^5 - 3x^2) dx = \frac{x^6}{6} - x^3 + C$

$\int_{-2}^{0} (x^5 - 3x^2) dx = \left[\frac{x^6}{6} - x^3\right]_{-2}^{0} = (0 - 0) - \left(\frac{64}{6} - (-8)\right) = 0 - \left(\frac{32}{3} + 8\right) = -\left(\frac{32}{3} + \frac{24}{3}\right) = -\frac{56}{3} \approx -18.667$

#### Численные методы:

Параметры:
- Интервал: [-2, 0]
- Длина интервала: 2
- Шаг: 0.2
- Количество узлов: 11

## Задание 2. Составить программу для вычисления интеграла из задания 1 по обобщенным формулам прямоугольников, трапеций и Симпсона.

На основе выполнения задания 1, составим программу, обобщающую все методы решения для всех трех интегралов:

```python
import numpy as np
from scipy import integrate
import math

print("\n" + "="*60)
print("Задание 2")
print("="*60)

# Функции для интегрирования
def f1(x):
    return np.sqrt(9*x - 1)

def f2(x):
    return (1 - 2*x)**4

def f3(x):
    return x**5 - 3*x**2

# Аналитические решения
def analytical_f1(a, b):
    return (2/27) * (9*b - 1)**(3/2) - (2/27) * (9*a - 1)**(3/2)

def analytical_f2(a, b):
    return (-1/10)*(1 - 2*b)**5 - (-1/10)*(1 - 2*a)**5

def analytical_f3(a, b):
    return (b**6/6 - b**3) - (a**6/6 - a**3)

def numerical_integration(f, a, b, n=11):
    """
    Вычисление интеграла тремя методами

    Parameters:
    f - функция
    a, b - пределы интегрирования
    n - количество узлов
    """

    # Проверка на четность для метода Симпсона
    if n % 2 == 0:
        n += 1

    x = np.linspace(a, b, n)
    y = f(x)
    h = (b - a) / (n - 1)

    # Метод прямоугольников (средних)
    rect = h * np.sum(y[:-1])

    # Метод трапеций
    trap = (h/2) * (y[0] + 2*np.sum(y[1:-1]) + y[-1])

    # Метод Симпсона
    simp = (h/3) * (y[0] + 4*np.sum(y[1:-1:2]) + 2*np.sum(y[2:-1:2]) + y[-1])

    return rect, trap, simp

# Тестируем на всех трех интегралах
methods = ["Прямоугольники", "Трапеции", "Симпсон"]

print("\nРезультаты вычислений:")
print("="*70)

# Интеграл 1
analytical1 = analytical_f1(1, 5)
rect1, trap1, simp1 = numerical_integration(f1, 1, 5, 11)
scipy1, _ = integrate.quad(f1, 1, 5)

print(f"\n∫₁⁵ √(9x - 1) dx:")
print(f"Аналитическое: {analytical1:.6f}")
print(f"Прямоугольники: {rect1:.6f} (погрешность: {abs(rect1 - analytical1):.2e})")
print(f"Трапеции:       {trap1:.6f} (погрешность: {abs(trap1 - analytical1):.2e})")
print(f"Симпсон:        {simp1:.6f} (погрешность: {abs(simp1 - analytical1):.2e})")
print(f"Scipy:          {scipy1:.6f}")

# Интеграл 2
analytical2 = analytical_f2(2, 3)
rect2, trap2, simp2 = numerical_integration(f2, 2, 3, 11)
scipy2, _ = integrate.quad(f2, 2, 3)

print(f"\n∫₂³ (1 - 2x)⁴ dx:")
print(f"Аналитическое: {analytical2:.6f}")
print(f"Прямоугольники: {rect2:.6f} (погрешность: {abs(rect2 - analytical2):.2e})")
print(f"Трапеции:       {trap2:.6f} (погрешность: {abs(trap2 - analytical2):.2e})")
print(f"Симпсон:        {simp2:.6f} (погрешность: {abs(simp2 - analytical2):.2e})")
print(f"Scipy:          {scipy2:.6f}")

# Интеграл 3
analytical3 = analytical_f3(-2, 0)
rect3, trap3, simp3 = numerical_integration(f3, -2, 0, 11)
scipy3, _ = integrate.quad(f3, -2, 0)

print(f"\n∫₋₂⁰ (x⁵ - 3x²) dx:")
print(f"Аналитическое: {analytical3:.6f}")
print(f"Прямоугольники: {rect3:.6f} (погрешность: {abs(rect3 - analytical3):.2e})")
print(f"Трапеции:       {trap3:.6f} (погрешность: {abs(trap3 - analytical3):.2e})")
print(f"Симпсон:        {simp3:.6f} (погрешность: {abs(simp3 - analytical3):.2e})")
print(f"Scipy:          {scipy3:.6f}")

# Сравнительная таблица
print("\n" + "="*70)
print("Сравнительная таблица погрешностей")
print("="*70)
print(f"{'Метод':<15} {'Интеграл 1':<12} {'Интеграл 2':<12} {'Интеграл 3':<12}")
print("-"*70)

errors1 = [abs(rect1 - analytical1), abs(trap1 - analytical1), abs(simp1 - analytical1)]
errors2 = [abs(rect2 - analytical2), abs(trap2 - analytical2), abs(simp2 - analytical2)]
errors3 = [abs(rect3 - analytical3), abs(trap3 - analytical3), abs(simp3 - analytical3)]

for i, method in enumerate(methods):
    print(f"{method:<15} {errors1[i]:<12.2e} {errors2[i]:<12.2e} {errors3[i]:<12.2e}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
import math

# Настройка стиля графиков
plt.style.use('default')
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Визуализация интегралов и построение графиков', fontsize=16, fontweight='bold')

# Функции для интегрирования
def f1(x):
    return np.sqrt(9*x - 1)

def f2(x):
    return (1 - 2*x)**4

def f3(x):
    return x**5 - 3*x**2

# Аналитические решения
def analytical_f1(a, b):
    return (2/27) * (9*b - 1)**(3/2) - (2/27) * (9*a - 1)**(3/2)

def analytical_f2(a, b):
    return (-1/10)*(1 - 2*b)**5 - (-1/10)*(1 - 2*a)**5

def analytical_f3(a, b):
    return (b**6/6 - b**3) - (a**6/6 - a**3)

# Параметры для графиков
x1_fine = np.linspace(0.8, 5.2, 200)
x2_fine = np.linspace(1.8, 3.2, 200)
x3_fine = np.linspace(-2.2, 0.2, 200)

# Узлы для численного интегрирования
x1_nodes = np.linspace(1, 5, 11)
x2_nodes = np.linspace(2, 3, 11)
x3_nodes = np.linspace(-2, 0, 11)

y1_nodes = f1(x1_nodes)
y2_nodes = f2(x2_nodes)
y3_nodes = f3(x3_nodes)

# Вычисление аналитических значений
analytical_val1 = analytical_f1(1, 5)
analytical_val2 = analytical_f2(2, 3)
analytical_val3 = analytical_f3(-2, 0)

# ИНТЕГРАЛ 1: ∫₁⁵ √(9x - 1) dx
ax1 = axes[0, 0]
ax1.plot(x1_fine, f1(x1_fine), 'b-', linewidth=2, label='f(x) = √(9x - 1)')
ax1.fill_between(x1_nodes, 0, f1(x1_nodes), alpha=0.3, color='red', label='Площадь интеграла')
ax1.scatter(x1_nodes, y1_nodes, color='red', s=30, zorder=5)
ax1.axvline(x=1, color='green', linestyle='--', alpha=0.7, label='Пределы интегрирования')
ax1.axvline(x=5, color='green', linestyle='--', alpha=0.7)
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.set_title(f'∫₁⁵ √(9x - 1) dx = {analytical_val1:.2f}')
ax1.grid(True, alpha=0.3)
ax1.legend()
ax1.set_xlim(0.8, 5.2)

# ИНТЕГРАЛ 2: ∫₂³ (1 - 2x)⁴ dx
ax2 = axes[0, 1]
ax2.plot(x2_fine, f2(x2_fine), 'b-', linewidth=2, label='f(x) = (1 - 2x)⁴')
ax2.fill_between(x2_nodes, 0, f2(x2_nodes), alpha=0.3, color='red', label='Площадь интеграла')
ax2.scatter(x2_nodes, y2_nodes, color='red', s=30, zorder=5)
ax2.axvline(x=2, color='green', linestyle='--', alpha=0.7, label='Пределы интегрирования')
ax2.axvline(x=3, color='green', linestyle='--', alpha=0.7)
ax2.set_xlabel('x')
ax2.set_ylabel('f(x)')
ax2.set_title(f'∫₂³ (1 - 2x)⁴ dx = {analytical_val2:.1f}')
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.set_xlim(1.8, 3.2)

# ИНТЕГРАЛ 3: ∫₋₂⁰ (x⁵ - 3x²) dx
ax3 = axes[0, 2]
ax3.plot(x3_fine, f3(x3_fine), 'b-', linewidth=2, label='f(x) = x⁵ - 3x²')
ax3.fill_between(x3_nodes, 0, f3(x3_nodes), alpha=0.3, color='red', label='Площадь интеграла')
ax3.scatter(x3_nodes, y3_nodes, color='red', s=30, zorder=5)
ax3.axvline(x=-2, color='green', linestyle='--', alpha=0.7, label='Пределы интегрирования')
ax3.axvline(x=0, color='green', linestyle='--', alpha=0.7)
ax3.set_xlabel('x')
ax3.set_ylabel('f(x)')
ax3.set_title(f'∫₋₂⁰ (x⁵ - 3x²) dx = {analytical_val3:.1f}')
ax3.grid(True, alpha=0.3)
ax3.legend()
ax3.set_xlim(-2.2, 0.2)

# Визуализация методов интегрирования для первого интеграла
ax4 = axes[1, 0]
x_demo = np.linspace(1, 5, 6)
y_demo = f1(x_demo)
h_demo = (5-1)/5

# Метод прямоугольников
ax4.step(x_demo[:-1], y_demo[:-1], where='post', color='orange', linewidth=2, label='Метод прямоугольников')
ax4.bar(x_demo[:-1], y_demo[:-1], width=h_demo, alpha=0.3, color='orange', align='edge')

# Метод трапеций
for i in range(len(x_demo)-1):
    ax4.fill([x_demo[i], x_demo[i], x_demo[i+1], x_demo[i+1]],
             [0, y_demo[i], y_demo[i+1], 0], alpha=0.2, color='green')

ax4.plot(x_demo, y_demo, 'ro-', linewidth=2, label='Метод трапеций')
ax4.plot(x1_fine, f1(x1_fine), 'b-', linewidth=1, alpha=0.5, label='Точная функция')
ax4.set_xlabel('x')
ax4.set_ylabel('f(x)')
ax4.set_title('Методы прямоугольников и трапеций')
ax4.grid(True, alpha=0.3)
ax4.legend()

# Визуализация методов Симпсона для первого интеграла
ax5 = axes[1, 1]
x_simp = np.linspace(1, 5, 11)
y_simp = f1(x_simp)

# Параболические аппроксимации
for i in range(0, len(x_simp)-2, 2):
    x_parab = np.linspace(x_simp[i], x_simp[i+2], 50)
    # Квадратичная интерполяция через три точки
    x0, x1, x2 = x_simp[i], x_simp[i+1], x_simp[i+2]
    y0, y1, y2 = y_simp[i], y_simp[i+1], y_simp[i+2]

    # Коэффициенты параболы
    a = ((y2-y0)/(x2-x0) - (y1-y0)/(x1-x0)) / (x2-x1)
    b = (y1-y0)/(x1-x0) - a*(x1+x0)
    c = y0 - a*x0**2 - b*x0

    y_parab = a*x_parab**2 + b*x_parab + c
    ax5.plot(x_parab, y_parab, 'g--', alpha=0.7, linewidth=1)
    ax5.fill_between(x_parab, 0, y_parab, alpha=0.2, color='green')

ax5.plot(x_simp, y_simp, 'ro-', linewidth=1, markersize=4, label='Узлы Симпсона')
ax5.plot(x1_fine, f1(x1_fine), 'b-', linewidth=2, label='Точная функция')
ax5.set_xlabel('x')
ax5.set_ylabel('f(x)')
ax5.set_title('Метод Симпсона (параболические сегменты)')
ax5.grid(True, alpha=0.3)
ax5.legend()

# Сравнение точности методов
ax6 = axes[1, 2]
methods = ['Прямоугольники', 'Трапеции', 'Симпсон']

# Вычисление реальных погрешностей
def calculate_errors():
    # Для интеграла 1
    rect1, trap1, simp1 = numerical_integration(f1, 1, 5, 11)
    error1_rect = abs(rect1 - analytical_val1)
    error1_trap = abs(trap1 - analytical_val1)
    error1_simp = abs(simp1 - analytical_val1)

    # Для интеграла 2
    rect2, trap2, simp2 = numerical_integration(f2, 2, 3, 11)
    error2_rect = abs(rect2 - analytical_val2)
    error2_trap = abs(trap2 - analytical_val2)
    error2_simp = abs(simp2 - analytical_val2)

    # Для интеграла 3
    rect3, trap3, simp3 = numerical_integration(f3, -2, 0, 11)
    error3_rect = abs(rect3 - analytical_val3)
    error3_trap = abs(trap3 - analytical_val3)
    error3_simp = abs(simp3 - analytical_val3)

    return ([error1_rect, error1_trap, error1_simp],
            [error2_rect, error2_trap, error2_simp],
            [error3_rect, error3_trap, error3_simp])

def numerical_integration(f, a, b, n=11):
    if n % 2 == 0:
        n += 1
    x = np.linspace(a, b, n)
    y = f(x)
    h = (b - a) / (n - 1)
    rect = h * np.sum(y[:-1])
    trap = (h/2) * (y[0] + 2*np.sum(y[1:-1]) + y[-1])
    simp = (h/3) * (y[0] + 4*np.sum(y[1:-1:2]) + 2*np.sum(y[2:-1:2]) + y[-1])
    return rect, trap, simp

errors1, errors2, errors3 = calculate_errors()

x_pos = np.arange(len(methods))
width = 0.25

ax6.bar(x_pos - width, errors1, width, label='∫₁⁵ √(9x-1)dx', alpha=0.8)
ax6.bar(x_pos, errors2, width, label='∫₂³ (1-2x)⁴dx', alpha=0.8)
ax6.bar(x_pos + width, errors3, width, label='∫₋₂⁰ (x⁵-3x²)dx', alpha=0.8)

ax6.set_xlabel('Методы интегрирования')
ax6.set_ylabel('Абсолютная погрешность')
ax6.set_title('Сравнение точности методов')
ax6.set_xticks(x_pos)
ax6.set_xticklabels(methods)
ax6.legend()
ax6.grid(True, alpha=0.3, axis='y')
ax6.set_yscale('log')  # Логарифмическая шкала для лучшего отображения

plt.tight_layout()
plt.show()
```

## Анализ графиков

### 1. Графики функций и площадей интегралов
- **Синяя линия**: точная функция
- **Красная заливка**: площадь под кривой (значение интеграла)
- **Красные точки**: узлы численного интегрирования
- **Зеленые пунктирные линии**: пределы интегрирования

### 2. Визуализация методов интегрирования
- **Метод прямоугольников**: ступенчатая аппроксимация
- **Метод трапеций**: линейная аппроксимация между узлами  
- **Метод Симпсона**: параболическая аппроксимация

### 3. Сравнение точности методов
- **Метод Симпсона** показывает наивысшую точность благодаря параболической аппроксимации
- **Метод трапеций** дает хороший компромисс между точностью и простотой
- **Метод прямоугольников** имеет наибольшую погрешность

### 4. Особенности вычисленных интегралов

#### Интеграл 1: ∫₁⁵ √(9x - 1) dx
- **Функция**: монотонно возрастающая корневая функция
- **Особенности**: плавное изменение, хорошо аппроксимируется всеми методами
- **Точность**: все методы показывают хорошие результаты благодаря гладкости функции

#### Интеграл 2: ∫₂³ (1 - 2x)⁴ dx  
- **Функция**: полином 4-й степени
- **Особенности**: значительное изменение производной на интервале
- **Точность**: метод Симпсона особенно эффективен для полиномиальных функций

#### Интеграл 3: ∫₋₂⁰ (x⁵ - 3x²) dx
- **Функция**: полином 5-й степени с отрицательным значением интеграла
- **Особенности**: функция меняет знак, площадь ниже оси OX
- **Точность**: методы хорошо справляются с отрицательными значениями

### 5. Зависимость от количества узлов
- С увеличением количества узлов погрешность уменьшается
- Метод Симпсона сходится быстрее всех благодаря параболической аппроксимации
- Для достижения высокой точности достаточно 11-15 узлов
- Метод прямоугольников требует больше узлов для сравнимой точности

### 6. Рекомендации по выбору метода
- **Для гладких функций**: метод Симпсона (высокая точность)
- **Для быстрых расчетов**: метод трапеций (хорошая точность при простоте)
- **Для монотонных функций**: все методы работают удовлетворительно
- **Для функций с быстрыми изменениями**: предпочтительнее метод Симпсона
